**Date Time functions**

```
d_parser = lambda x: pd.datetime.strptime(x, '%Y-%m-%d %I-%p')
url='https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Pandas/10-Datetime-Timeseries/ETH_1h.csv'
time_df = pd.read_csv(url, parse_dates=['Date'], date_parser=d_parser)
```

```
time_df.head()
```

```
time_df.loc[0, 'Date'].day_name()
```

```
time_df['Date'].dt.day_name() #similar like str
```

```
time_df['DayOfWeek'] = time_df['Date'].dt.day_name()
```

```
time_df['Date'].min()
```

```
time_df['Date'].max()
```

```
filt = (time_df['Date'] >= pd.to_datetime('2019-01-01')) & (time_df['Date'] < pd.to_datetime('2020-01-01'))
time_df.loc[filt]
```

```

time_df.set_index('Date', inplace=True)
```

```
time_df['2019']
```
